```python
import pandas as pd
df = pd.read_csv("/content/Video_Games_Sales_as_at_22_Dec_2016.csv")
```

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
df.head(20)
```

```python
df["Year_of_Release"] = df["Year_of_Release"].astype("int64")
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df.rename(columns={"Name":"Nome",	"Platform":"Plataforma",	"Year_of_Release":"Ano_lancamento",	"Genre":"Genero",	"Publisher":"Publicadora", "NA_Sales":"NA_Vendas",	"EU_Sales":"EU_Vendas",	"JP_Sales":"JP_Vendas",	"Other_Sales":"Outras_Vendas",	"Global_Sales":"Vendas_Globais", "Critic_Score":"Nota_Critica",	"Critic_Count":"Contagem_Criticas",	"User_Score":"Nota_Usuario",	"User_Count":"Contagem_Usuario",	"Developer":"Desenvolvedora",	"Rating":"Nota"}, inplace=True)
```

```python
df.head(5)
```

```python
df["Plataforma"].value_counts().plot.bar(title="Número de jogos x Plataforma", color="blue", rot=0, figsize= (11,5))
plt.xlabel("Plataforma")
plt.ylabel("Numero de jogos")
fig = plt.gcf()
plt.savefig('jogos_x_plat.png', format='png')
plt.show()
```

```python
top_10= df[["Nome","Vendas_Globais"]].head(10).set_index("Nome").sort_values("Vendas_Globais", ascending=True)
top_10.plot.barh(title= "Top 10 jogos mais vendidos", color="blue", rot=0, figsize= (18,5))
plt.xlabel("Vendas (Em milhões de dólares)")
plt.ylabel("Jogos")
fig = plt.gcf()
plt.savefig('Top10.png', format='png')
plt.show();
```
